```python
import pandas as pd
import numpy as np
Df = pd.read_csv("melb_data.csv")
Df.head()
```

```python
def random_sample_imputation(df):
    coloumns_with_missing_values = df.columns [df.isna().any()].tolist()
    for var in coloumns_with_missing_values:
       random_sample_df = df[var].dropna().sample(df[var].isnull().sum(),
                                                  random_state=0)
       random_sample_df.index = df[
            df[var].isnull()].index

    # replace the NA
       df.loc[df[var].isnull(), var] = random_sample_df
 
    return df
```

```python
Df = pd.read_csv("melb_data.csv")
df_imputed = random_sample_imputation(Df)
df_imputed
```
